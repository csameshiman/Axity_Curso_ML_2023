# K Means Clustering - Ejemplo - Blob

**Contexto**  
Generaremos un conjunto de datos aleatorio.

**Contenido**  
El conjunto de datos se generará con la utilería make_blobs de scikit-learn: [make_blobs](https://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_blobs.html#sklearn.datasets.make_blobs).  

**Planteamiento del problema**  
Se busca identificar los grupos que pudieran existir dentro de los datos.

```python
# Importar librerias
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

## Crear el conjunto de datos

```python
data = make_blobs(n_samples=200, n_features=2, 
                  centers=4, cluster_std=1.8, random_state=42)
```

```python
# Visualizar los datos
plt.scatter(data[0][:,0],data[0][:,1],c=data[1])
```

## Modelado

```python
# Entrenamiento
model = KMeans(n_clusters=3, n_init=10)
model.fit(data[0])
```

```python
# Resultados
model.cluster_centers_
```

```python
model.labels_
```

## Visualización

```python
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True,figsize=(10,6))
ax1.set_title('K Means')
ax1.scatter(data[0][:,0],data[0][:,1],c=model.labels_)
ax2.set_title("Original")
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1])
```

Nota: los colores no tienen relación entre los 2 gráficos
